##Entrenamiento y generar tabla de resultados

```python
ruta = '/content/drive/MyDrive/Doctorado/Articulos/04. Sistema de indeficación/Modelos/'

X = df_all.loc[:,df_all.columns!='peso'].values
y = df_all.loc[:,'peso'].values

n_estimators = 10
cantidad = 35

def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

resultado = []
for i in range(cantidad):
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = i)

  start_time = time()
  
  modelo = RandomForestRegressor(n_estimators = n_estimators, random_state = i,max_depth=100)
  modelo.fit(X_train, y_train)
  
  elapsed_time = time() - start_time

  predictions = modelo.predict(X_test)
  errors = abs(predictions - y_test)
  resultado.append(["RandomForest_Regressor",mean_absolute_error(y_test, predictions),mean_squared_error(y_test, predictions, squared=False),mean_absolute_percentage_error(y_test, predictions),r2_score(y_test, predictions),i,elapsed_time])
  df_r = pd.DataFrame(resultado)
  df_r = df_r.rename(columns={0:'Modelo',1:'MAE',2:'MSE',3:'MAPE',4:'r2',5:'random_state',6:'time'})
  df_r.to_csv(ruta+'RandomForest_Regressor.csv')

resultado = []
for i in range(cantidad):
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = i)

  start_time = time()

  modelo = DecisionTreeRegressor(random_state = i)
  modelo.fit(X_train, y_train)

  elapsed_time = time() - start_time

  predictions = modelo.predict(X_test)
  errors = abs(predictions - y_test)
  resultado.append(["DecisionTree_Regressor",mean_absolute_error(y_test, predictions),mean_squared_error(y_test, predictions, squared=False),mean_absolute_percentage_error(y_test, predictions),r2_score(y_test, predictions),i,elapsed_time])
  df_r = pd.DataFrame(resultado)
  df_r = df_r.rename(columns={0:'Modelo',1:'MAE',2:'MSE',3:'MAPE',4:'r2',5:'random_state',6:'time'})
  df_r.to_csv(ruta+'DecisionTree_Regressor.csv')

resultado = []
for i in range(cantidad):
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = i)

  start_time = time()

  modelo = LinearRegression()
  modelo.fit(X_train, y_train)

  elapsed_time = time() - start_time

  predictions = modelo.predict(X_test)
  errors = abs(predictions - y_test)
  resultado.append(["Linear_Regression",mean_absolute_error(y_test, predictions),mean_squared_error(y_test, predictions, squared=False),mean_absolute_percentage_error(y_test, predictions),r2_score(y_test, predictions),i,elapsed_time])
  df_r = pd.DataFrame(resultado)
  df_r = df_r.rename(columns={0:'Modelo',1:'MAE',2:'MSE',3:'MAPE',4:'r2',5:'random_state',6:'time'})
  df_r.to_csv(ruta+'Linear_Regression.csv')

resultado = []
for i in range(cantidad):
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = i)

  start_time = time()

  modelo = GradientBoostingRegressor(random_state = i)
  modelo.fit(X_train, y_train)

  elapsed_time = time() - start_time

  predictions = modelo.predict(X_test)
  errors = abs(predictions - y_test)
  resultado.append(["GradientBoostingRegressor",mean_absolute_error(y_test, predictions),mean_squared_error(y_test, predictions, squared=False),mean_absolute_percentage_error(y_test, predictions),r2_score(y_test, predictions),i,elapsed_time])
  df_r = pd.DataFrame(resultado)
  df_r = df_r.rename(columns={0:'Modelo',1:'MAE',2:'MSE',3:'MAPE',4:'r2',5:'random_state',6:'time'})
  df_r.to_csv(ruta+'GradientBoostingRegressor.csv')

resultado = []
for i in range(cantidad):
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = i)

  start_time = time()

  modelo = AdaBoostRegressor(random_state = i, n_estimators=n_estimators)
  modelo.fit(X_train, y_train)

  elapsed_time = time() - start_time

  predictions = modelo.predict(X_test)
  errors = abs(predictions - y_test)
  resultado.append(["AdaBoostRegressor",mean_absolute_error(y_test, predictions),mean_squared_error(y_test, predictions, squared=False),mean_absolute_percentage_error(y_test, predictions),r2_score(y_test, predictions),i,elapsed_time])
  df_r = pd.DataFrame(resultado)
  df_r = df_r.rename(columns={0:'Modelo',1:'MAE',2:'MSE',3:'MAPE',4:'r2',5:'random_state',6:'time'})
  df_r.to_csv(ruta+'AdaBoostRegressor.csv')

resultado = []
for i in range(cantidad):
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = i)

  start_time = time()

  modelo = KNeighborsRegressor(n_neighbors=2)
  modelo.fit(X_train, y_train)

  elapsed_time = time() - start_time

  predictions = modelo.predict(X_test)
  errors = abs(predictions - y_test)
  resultado.append(["KNeighborsRegressor",mean_absolute_error(y_test, predictions),mean_squared_error(y_test, predictions, squared=False),mean_absolute_percentage_error(y_test, predictions),r2_score(y_test, predictions),i,elapsed_time])
  df_r = pd.DataFrame(resultado)
  df_r = df_r.rename(columns={0:'Modelo',1:'MAE',2:'MSE',3:'MAPE',4:'r2',5:'random_state',6:'time'})
  df_r.to_csv(ruta+'KNeighborsRegressor.csv')

resultado = []
for i in range(cantidad):
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = i)

  start_time = time()

  modelo = MLPRegressor(random_state=i, max_iter=500)
  modelo.fit(X_train, y_train)

  elapsed_time = time() - start_time

  predictions = modelo.predict(X_test)
  errors = abs(predictions - y_test)
  resultado.append(["MLPRegressor",mean_absolute_error(y_test, predictions),mean_squared_error(y_test, predictions, squared=False),mean_absolute_percentage_error(y_test, predictions),r2_score(y_test, predictions),i,elapsed_time])
  df_r = pd.DataFrame(resultado)
  df_r = df_r.rename(columns={0:'Modelo',1:'MAE',2:'MSE',3:'MAPE',4:'r2',5:'random_state',6:'time'})
  df_r.to_csv(ruta+'MLPRegressor.csv')
```
