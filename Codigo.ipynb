```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import  train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

df = pd.read_csv('/content/data_full.csv')
df = df.drop(['Unnamed: 0'],axis=1)

df2 = (df[df['NELLORE']==1])
df2 = df2.drop(['BON','CEBU','NELLORE','ROMO'],axis=1)

df2=df2.reset_index()
df2 = df2.drop(['index'],axis=1)

df2['PESO_PROMEDIO'] = 0
df_peso_p = df2.groupby('MES').mean()['PESO SIMULADO']

for j in range(len(df_peso_p)):
  df2.iloc[df2[(df2['MES']==df_peso_p.index[j])].index,2]=df2.groupby('MES').mean()['PESO SIMULADO'][df_peso_p.index[j]]

X = df2.loc[:,df2.columns!='PESO SIMULADO'].values
y = df2.loc[:,'PESO SIMULADO'].values

n_estimators = 100
cantidad = 35

def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

resultado = []
for i in range(cantidad):
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = i)
  modelo = RandomForestRegressor(n_estimators = n_estimators, random_state = i,max_depth=100)
  modelo.fit(X_train, y_train)
  predictions = modelo.predict(X_test)
  errors = abs(predictions - y_test)
  resultado.append(["RandomForest_Regressor",mean_absolute_error(y_test, predictions),mean_squared_error(y_test, predictions, squared=False),mean_absolute_percentage_error(y_test, predictions),r2_score(y_test, predictions),i])
  df_r = pd.DataFrame(resultado)
  df_r = df_r.rename(columns={0:'Modelo',1:'MAE',2:'MSE',3:'MAPE',4:'r2',5:'random_state'})
  df_r.to_csv('RandomForest_Regressor.csv')

resultado = []
for i in range(cantidad):
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = i)
  modelo = DecisionTreeRegressor(random_state = i)
  modelo.fit(X_train, y_train)
  predictions = modelo.predict(X_test)
  errors = abs(predictions - y_test)
  resultado.append(["DecisionTree_Regressor",mean_absolute_error(y_test, predictions),mean_squared_error(y_test, predictions, squared=False),mean_absolute_percentage_error(y_test, predictions),r2_score(y_test, predictions),i])
  df_r = pd.DataFrame(resultado)
  df_r = df_r.rename(columns={0:'Modelo',1:'MAE',2:'MSE',3:'MAPE',4:'r2',5:'random_state'})
  df_r.to_csv('DecisionTree_Regressor.csv')

resultado = []
for i in range(cantidad):
  X_train, X_test, y_train, y_test = train_test_split
(X, y, test_size = 0.2, random_state = i)
  modelo = LinearRegression()
  modelo.fit(X_train, y_train)
  predictions = modelo.predict(X_test)
  errors = abs(predictions - y_test)
  resultado.append(["Linear_Regression",mean_absolute_error(y_test, predictions),mean_squared_error(y_test, predictions, squared=False),mean_absolute_percentage_error(y_test, predictions),r2_score(y_test, predictions),i])
  df_r = pd.DataFrame(resultado)
  df_r = df_r.rename(columns={0:'Modelo',1:'MAE',2:'MSE',3:'MAPE',4:'r2',5:'random_state'})
  df_r.to_csv('Linear_Regression.csv')

resultado = []
for i in range(cantidad):
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = i)
  modelo = GradientBoostingRegressor(random_state = i)
  modelo.fit(X_train, y_train)
  predictions = modelo.predict(X_test)
  errors = abs(predictions - y_test)
  resultado.append(["GradientBoostingRegressor",mean_absolute_error(y_test, predictions),mean_squared_error(y_test, predictions, squared=False),mean_absolute_percentage_error(y_test, predictions),r2_score(y_test, predictions),i])
  df_r = pd.DataFrame(resultado)
  df_r = df_r.rename(columns={0:'Modelo',1:'MAE',2:'MSE',3:'MAPE',4:'r2',5:'random_state'})
  df_r.to_csv('GradientBoostingRegressor.csv')

resultado = []
for i in range(cantidad):
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = i)
  modelo = AdaBoostRegressor(random_state = i, n_estimators=n_estimators)
  modelo.fit(X_train, y_train)
  predictions = modelo.predict(X_test)
  errors = abs(predictions - y_test)
  resultado.append(["AdaBoostRegressor",mean_absolute_error(y_test, predictions),mean_squared_error(y_test, predictions, squared=False),mean_absolute_percentage_error(y_test, predictions),r2_score(y_test, predictions),i])
  df_r = pd.DataFrame(resultado)
  df_r = df_r.rename(columns={0:'Modelo',1:'MAE',2:'MSE',3:'MAPE',4:'r2',5:'random_state'})
  df_r.to_csv('AdaBoostRegressor.csv')

resultado = []
for i in range(cantidad):
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = i)
  modelo = KNeighborsRegressor(n_neighbors=2)
  modelo.fit(X_train, y_train)
  predictions = modelo.predict(X_test)
  errors = abs(predictions - y_test)
  resultado.append(["KNeighborsRegressor",mean_absolute_error(y_test, predictions),mean_squared_error(y_test, predictions, squared=False),mean_absolute_percentage_error(y_test, predictions),r2_score(y_test, predictions),i])
  df_r = pd.DataFrame(resultado)
  df_r = df_r.rename(columns={0:'Modelo',1:'MAE',2:'MSE',3:'MAPE',4:'r2',5:'random_state'})
  df_r.to_csv('KNeighborsRegressor.csv')

resultado = []
for i in range(cantidad):
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = i)
  modelo = MLPRegressor(random_state=i, max_iter=500)
  modelo.fit(X_train, y_train)
  predictions = modelo.predict(X_test)
  errors = abs(predictions - y_test)
  resultado.append(["MLPRegressor",mean_absolute_error(y_test, predictions),mean_squared_error(y_test, predictions, squared=False),mean_absolute_percentage_error(y_test, predictions),r2_score(y_test, predictions),i])
  df_r = pd.DataFrame(resultado)
  df_r = df_r.rename(columns={0:'Modelo',1:'MAE',2:'MSE',3:'MAPE',4:'r2',5:'random_state'})
  df_r.to_csv('MLPRegressor.csv')


from sklearn.svm import LinearSVR
resultado = []
for i in range(cantidad):
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = i)
  modelo = LinearSVR(random_state=i)
  modelo.fit(X_train, y_train)
  predictions = modelo.predict(X_test)
  errors = abs(predictions - y_test)
  resultado.append(["LinearSVR",mean_absolute_error(y_test, predictions),mean_squared_error(y_test, predictions, squared=False),mean_absolute_percentage_error(y_test, predictions),r2_score(y_test, predictions),i])
  df_r = pd.DataFrame(resultado)
  df_r = df_r.rename(columns={0:'Modelo',1:'MAE',2:'MSE',3:'MAPE',4:'r2',5:'random_state'})
  df_r.to_csv('LinearSVR.csv')
```
