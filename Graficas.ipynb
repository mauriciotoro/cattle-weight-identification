```python
d = df_pesaje[ (df_pesaje['raza']=="AC") & (df_pesaje['genero']=="Hembra") & (df_pesaje['peso']>=100)& (df_pesaje['peso']<=455) ]
d=d.reset_index()

data = df_all[ (df_all['raza_AC']==1) & (df_all['genero_Hembra']==1) ]
data = data.groupby(by=["dia"]).mean()
data['dia']=data.index

fig = plt.figure(figsize=(17,19))
ax = fig.add_subplot(221)

ax.set_ylabel('Weights', fontsize=14)
ax.set_xlabel('Days', fontsize=14)
data = data.rename(columns={'dia':'Days'})

#plt.title("Ideal weight curve versus real weight in AC breed, female", fontsize=14)
data.plot(kind='line',x='Days',y='peso', ax=ax, label='Ideal weight')
for i in range(len(d)):
  plt.text(d['dia'][i], d['peso'][i], '*', fontsize=12, color='black')
plt.grid(True)
```
